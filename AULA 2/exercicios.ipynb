```python
fig = plt.figure(figsize=(15, 8))
plt.rcParams['font.size'] = '20'

eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(df['data'], df['pressão'], color='b', lw=0.1, linestyle='--', marker='')


ultimo_ponto = len(df) - 1

eixo.annotate(
    f"{df['pressão'][ultimo_ponto]} hPa",  # Texto da anotação (valor da pressão no último ponto)
    (df['data'][ultimo_ponto], df['pressão'][ultimo_ponto]),  # Coordenadas para a anotação
    textcoords="offset points",  # Tipo de coordenadas para o texto
    xytext=(0, 10),  # Deslocamento do texto em relação ao ponto
    ha='center',  # Alinhamento horizontal do texto
    fontsize=15,  # Tamanho da fonte do texto
    color='black'  # Cor do texto
)

eixo.set_title('Pressão', fontsize=25)
eixo.legend(['Pressão'], loc='lower right', fontsize=15)
eixo.set_ylabel('Pressão (hPa)', fontsize=20)  # Adiciona unidade de hPa
eixo.set_xlabel('Data', fontsize=20)  # Alterado para 'Data' para corresponder aos dados reais

plt.grid(True)
plt.show()
```

```python

df['data'] = pd.to_datetime(df['data'])
periodo_2016 = df[(df['data'] >= '2016-05-01') & (df['data'] <= '2016-08-31')]
ano_2016 = df[df['data'].dt.year == 2016]
fig = plt.figure(figsize=(14, 10))

eixo1 = fig.add_axes([0, 0.5, 1, 0.5])
eixo1.plot(periodo_2016['data'], periodo_2016['temperatura'], color='red')
eixo1.set_title('Temperatura de 05-2016 a 08-2016', fontsize=20)
eixo1.set_ylabel('Temperatura (°C)', fontsize=15)
eixo1.grid(True)

eixo2 = fig.add_axes([0, 0, 1, 0.5], sharex=eixo1)
eixo2.plot(ano_2016['data'], ano_2016['temperatura'], color='blue')
eixo2.set_title('Temperatura ao Longo de 2016', fontsize=20)
eixo2.set_ylabel('Temperatura (°C)', fontsize=15)
eixo2.set_xlabel('Data', fontsize=15)
eixo2.grid(True)

plt.tight_layout()
plt.show()

```

```python


df['data'] = pd.to_datetime(df['data'])
periodo_2017 = df[(df['data'] >= '2017-05-01') & (df['data'] <= '2017-08-31')]
fig = plt.figure(figsize=(14, 10))

eixo_maior = fig.add_axes([0, 0, 1, 1])  
eixo_maior.plot(periodo_2017['data'], periodo_2017['temperatura'], color='red', label='Temperatura')
eixo_maior.set_title('Temperatura de 05-2017 a 08-2017', fontsize=20)
eixo_maior.set_ylabel('Temperatura (°C)', fontsize=15)
eixo_maior.grid(True)
eixo_maior.legend(loc='best', fontsize=12) 

eixo_menor = fig.add_axes([0.6, 0.1, 0.35, 0.35]) 
eixo_menor.plot(periodo_2017['data'], periodo_2017['humidade'], color='blue', label='Umidade')
eixo_menor.set_title('Umidade de 05-2017 a 08-2017', fontsize=15)
eixo_menor.set_ylabel('Umidade (%)', fontsize=10)
eixo_menor.set_xlabel('Data', fontsize=10)
eixo_menor.grid(True)
eixo_menor.legend(loc='upper left', fontsize=8, frameon=False) 

plt.tight_layout()
plt.show()



```
